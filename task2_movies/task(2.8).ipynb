```python
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.concat([pd.read_csv(f) for f in glob.glob('./user_results/part-*.csv')])

# 1. Top Users Table
top_users = df.sort_values('rating_count', ascending=False).head(10)
print("Top 10 Users by Rating Count:")
print(top_users[['userId', 'rating_count']].to_string(index=False))

# 2. User Distribution Chart
dist = df['user_type'].value_counts()

plt.figure(figsize=(8, 6))
bars = plt.bar(dist.index, dist.values, color=['#1f77b4', '#ff7f0e'])
plt.title("User Activity Distribution", pad=20)
plt.xlabel("User Type")
plt.ylabel("Number of Users")
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:,}',
             ha='center', va='bottom')

plt.tight_layout()
plt.savefig('user_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
