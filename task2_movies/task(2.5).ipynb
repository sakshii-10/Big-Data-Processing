```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Load data
df = pd.concat([pd.read_csv(f) for f in glob.glob('./genre_results/part-*.csv')])

# Create chart
plt.figure(figsize=(10,5))
ax = df.plot.bar(x='genre', y='rating_count', legend=False, color='#1f77b4')

# Customize
plt.title("Movie Ratings by Genre", pad=20, weight='bold')
plt.xlabel("Genre", labelpad=10)
plt.ylabel("Number of Ratings", labelpad=10)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.tight_layout()
plt.savefig('genre_distribution.png', dpi=300, bbox_inches='tight')
print("Saved: genre_distribution.png")
plt.show()
```
