```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

# Load data
df = pd.concat([pd.read_csv(f) for f in glob.glob('./yearly_results/part-*.csv')])

# Create line chart
plt.figure(figsize=(12, 6))
plt.plot(df['year'], df['total_ratings'], 
         marker='o', 
         linestyle='-', 
         color='#1f77b4',
         linewidth=2,
         markersize=8)

# Formatting
plt.title("Movie Ratings Over Time", fontsize=14, pad=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Ratings", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add data labels
for x, y in zip(df['year'], df['total_ratings']):
    plt.text(x, y, f"{y:,}", 
             ha='center', 
             va='bottom',
             fontsize=9)

plt.tight_layout()
plt.savefig('ratings_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```
